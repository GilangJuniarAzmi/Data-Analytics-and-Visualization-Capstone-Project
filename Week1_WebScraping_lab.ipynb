# **Extract information from the given web site**

We will extract the data from the below web site:

```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data we need to scrape is the name of the **programming language** and **average annual salary**.
It is a good idea to open the url in our web broswer and study the contents of the web page before we start to scrape.

Import the required libraries

```python
!pip install bs4
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests # this module helps us to download a web page
import pandas as pd
```

```python
data = requests.get(url).text # download the webpage at the url
soup = BeautifulSoup(data, 'html5lib') # create a soup object
```

Scrape the <code>Language name</code> and <code>annual average salary</code>.

```python
rows_data = []

for row in soup.find("tbody").find_all("tr"):
  col = row.find_all("td")
  # Ensure there are enough columns before trying to access specific indices
  language = col[1].text
  salary = col[3].text
  rows_data.append({'Language Name' : language, 'Average Salary' : salary})

# Create the DataFrame from the list of dictionaries
df = pd.DataFrame(rows_data)

# The first row scraped
if not df.empty:
    df.drop(index=df.index[0], axis=0, inplace=True)

print(df)
```

Save the scrapped data into a file named *popular-languages.csv*

```python
df.to_csv('popular-languages.csv')
```
